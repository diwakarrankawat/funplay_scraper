```python
from data import scraper
def read_urls_file():
    urls_data = {}
    with open('urls.txt', 'r') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            url, servers = line.split(',', maxsplit=1)
            urls_data[url.strip()] = [s.strip() for s in servers.split(',')]
    return urls_data

urls = read_urls_file()
urls
```

```python
dt = scraper.Data(urls)
```
